```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

url = "https://raw.githubusercontent.com/PhilChodrow/ml-notes/main/data/palmer-penguins/palmer-penguins.csv"
df = pd.read_csv(url)

#show table
df
```

```python
#Warm-Up part A

#group dataset based on sex and species to find mean mass

meanMassTable = df.groupby(['Species', 'Sex']).aggregate({'Body Mass (g)' : 'mean'}) #using dictionary to apply the mean function on the data
print(meanMassTable)
```

```python
#Warm-up part B

#Scatterplot of culmen length against culmen depth (bill dimensions), with the color of each point corresponding to the penguin species

#creating scatterplot 
sns.scatterplot(data = df, x = 'Culmen Length (mm)', y = 'Culmen Depth (mm)', hue = 'Species')

#labeling plot
plt.title('Culmen Length vs. Culmen Depth Based on Penguin Species')
plt.show()
```
