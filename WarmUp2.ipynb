**Graphing Decision Boundaries**

Part A:

We use the dot product equation $w_i x_i + w_j x_j = b$ and substitute the values for w and b:
$$(1) * x_i + -\dfrac 1 2 * x_j = \dfrac 1 2$$
We make the equation into $y = mx + b$ to plot it: 
$$ x_j = 2 * x_i - 1$$

```python
import numpy as np
import matplotlib.pyplot as plt

#making range for x
x_vals = np.linspace(-5, 10, num = 10)

#solving for x2
x2_vals = 2 * x_vals - 1

plt.plot(x_vals, x2_vals, color = 'pink')

plt.xlabel = ('x1')
plt.ylabel = ('x2')
plt.title = ("Graph of 1(x1) + (-1/2)x2 = 1/2")
plt.grid (True, linestyle = '--')
plt.show()

```

Your function should return 0 if dot product of w and x < b and 1 if the dot product of w and x >= b.

```python
import numpy as np
def linear_classify(x, w, b):
    if  np.dot(w, x) < b: 
        return 0
    elif (w@x) >= b: 
        return 1
    
#testing
    #should return 1 since dot prod > b
w = np.array([1, 2, 3, 4])
x = np.array([2, 6, -1, 0])
b = 3
    #should return 0 since dot prod < b
w1 = np.array([1, 2, 3, 4])
x1 = np.array([2, 6, -1, 0])
b1 = 30

w2 = np.array([2, -10, -5])
x2 = np.array([2, -1, 0])
b2 = 10

testing = linear_classify(w, x, b)
testing2 = linear_classify(w1, x1, b1)
testing3 = linear_classify(w2, x2, b2)

print('Test 1: ', testing)
print('Test 2: ', testing2)
print('Test 2: ', testing3)

```

Part C:

Make a sketch of the curve in the nonnegative quadrant defined by the equation given.

```python
import numpy as np
import matplotlib.pyplot as plt

#making range for x
x_vals = np.linspace(0, 10)
x1_val = np.sqrt(1 - 1/4 * (x_vals**2))

plt.plot(x_vals, x1_val, color = 'pink')
plt.xlabel = ('x')
plt.ylabel = ('x1')
plt.title = ('Curve in Nonnegative Quadrant')
plt.grid (True, linestyle = '--')
plt.show()
```
