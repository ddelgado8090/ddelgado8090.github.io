```python
import numpy as np
import matplotlib.pyplot as plt

def logistic_sigmoid (z):
    sig_x = 1 / (1 + np.exp(-z))
    return sig_x
z_val = np.linspace(-1, 1, 100)
sig_function = logistic_sigmoid(z_val)

plt.plot(z_val, sig_function)
plt.xlabel('z values')
plt.ylabel('sigmoid(z)')
plt.show()

```
